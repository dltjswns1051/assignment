```python
import tkinter as tk
from tkinter import messagebox
import csv
import os

file_path = "data.csv"

if not os.path.isfile(file_path):
    with open(file_path, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["User ID", "Username", "Email", "Birth Year"])

def insert_data():
    user_id = entry_user_id.get()
    username = entry_username.get()
    email = entry_email.get()
    birthyear = entry_birthyear.get()

    if not user_id or not username or not email or not birthyear:
        messagebox.showinfo("오류", "데이터 입력 중 오류가 발생함")
        return
    data = [user_id, username, email, birthyear]

    with open(file_path, 'a', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(data)

    messagebox.showinfo("알림", "데이터 입력 성공")

def search_data():
    user_id = entry_user_id.get()
    found = False

    for i in range(4):
        result_listboxes[i].delete(0, tk.END)

    with open(file_path, 'r', newline='') as csv_file:
        reader = csv.reader(csv_file)
        next(reader)

        for row in reader:
            if row[0] == user_id:
                found = True
                display_user_info(row)

    if not found:
        messagebox.showinfo("알림", "데이터가 없습니다")

def display_user_info(user_data):
    for i in range(4):
        result_listboxes[i].insert(tk.END, user_data[i])

window = tk.Tk()
window.title("GUI 데이터 입력")

# 입력 필드 프레임
input_frame = tk.Frame(window)
input_frame.pack()

# 아이디
entry_user_id = tk.Entry(input_frame)
entry_user_id.grid(row=0, column=0)

# 이름
entry_username = tk.Entry(input_frame)
entry_username.grid(row=0, column=1)

# 메일
entry_email = tk.Entry(input_frame)
entry_email.grid(row=0, column=2)

# 출생연도
entry_birthyear = tk.Entry(input_frame)
entry_birthyear.grid(row=0, column=3)

#입력
insert_button = tk.Button(input_frame, text="입력", command=insert_data)
insert_button.grid(row=1, column=0, columnspan=4)

#조회
search_button = tk.Button(input_frame, text="조회", command=search_data)
search_button.grid(row=2, column=0, columnspan=4)

# 네 개의 리스트박스
result_listboxes = [tk.Listbox(window, height=10, width=15, bg='yellow') for _ in range(4)]
for i in range(4):
    result_listboxes[i].pack(side=tk.LEFT, padx=10)

# 메인 루프 실행
window.mainloop()
```
